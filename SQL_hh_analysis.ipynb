1. Напишите запрос, который позволит узнать, какие работодатели находятся на первом и пятом месте по количеству вакансий.

```python
# текст запроса
query_5_1 = '''
WITH ranked_employers AS (
    SELECT 
        e.name AS employer,
        COUNT(v.id) AS vacancies_count,
        ROW_NUMBER() OVER (ORDER BY COUNT(v.id) DESC) AS rn
    FROM vacancies v
    JOIN employers e ON v.employer_id = e.id
    GROUP BY e.name
)
SELECT employer, vacancies_count
FROM ranked_employers
WHERE rn IN (1, 5);
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_5_1, connection)
df
```

2. Напишите запрос, который для каждого региона выведет количество работодателей и вакансий в нём.
Среди регионов, в которых нет вакансий, найдите тот, в котором наибольшее количество работодателей.


```python
# текст запроса
query_5_2 = '''
WITH employers_per_area AS (
       SELECT 
              e.area AS area_id,
              COUNT(*) AS employers_count
         FROM employers AS e
        GROUP BY e.area
),
vacancies_per_area AS (
       SELECT 
              v.area_id,
              COUNT(*) AS vacancies_count
         FROM vacancies AS v
        GROUP BY v.area_id
),
area_stats AS (
       SELECT 
              a.id AS area_id,
              a.name AS area_name,
              COALESCE(emp.employers_count, 0) AS employers_count,
              COALESCE(vac.vacancies_count, 0) AS vacancies_count
         FROM areas AS a
         LEFT JOIN employers_per_area AS emp
           ON emp.area_id = a.id
         LEFT JOIN vacancies_per_area AS vac
           ON vac.area_id = a.id
)
SELECT 
       area_name,
       employers_count,
       vacancies_count
  FROM area_stats
 ORDER BY vacancies_count ASC, employers_count DESC;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_5_2, connection)
df
```

```python
print(f'Наибольшее количество вакансий в стране: {df[df["vacancies_count"] == 0].iloc[0]["area_name"]}')
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в среднем указывают для каждого типа требуемого опыта (уникальное значение из поля `experience`).

    При решении задачи примите во внимание следующее:
    1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
    2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
    3. Если в расчётах участвует `null`, в результате он тоже даст `null` (посмотрите, что возвращает запрос `select 1 + null`). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит `null` на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

    Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до **целого числа**.

```python
# текст запроса
query_6_6 = f'''
SELECT
    experience,
    ROUND(
        AVG(
            (COALESCE(salary_from, 0) + COALESCE(salary_to, 0)) / 
            CASE 
                WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL THEN 2
                ELSE 1
            END
        )
    )::integer AS avg_salary
FROM vacancies
WHERE 
    (
        LOWER(name) LIKE '%data scientist%' OR
        LOWER(name) LIKE '%data science%' OR
        LOWER(name) LIKE '%исследователь данных%' OR
        LOWER(name) LIKE '%ml%' AND LOWER(name) NOT LIKE '%html%' OR
        LOWER(name) LIKE '%machine learning%' OR
        LOWER(name) LIKE '%машинн%обучен%'
    )
    AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
GROUP BY experience
ORDER BY experience;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_6_6, connection)
df
```

```python
# текст запроса
query_7_2 = f'''
-- Компании с наибольшим количеством уникальных вакансий
SELECT e.name AS employer, COUNT(*) AS vacancies_count
FROM vacancies v
JOIN employers e ON v.employer_id = e.id
GROUP BY e.name
ORDER BY vacancies_count DESC
LIMIT 20;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_2, connection)
df
```

```python
# текст запроса
query_7_3 = f'''
-- Средняя зарплата и количество вакансий по отраслям
SELECT 
    i.name AS industry,
    ROUND(AVG(
        (COALESCE(v.salary_from, 0) + COALESCE(v.salary_to, 0)) / 
        CASE 
            WHEN v.salary_from IS NOT NULL AND v.salary_to IS NOT NULL THEN 2 
            ELSE 1 
        END
    )) AS avg_salary,
    COUNT(v.id) AS vacancies_count
FROM vacancies v
JOIN employers e ON v.employer_id = e.id
LEFT JOIN employers_industries ei ON e.id = ei.employer_id
LEFT JOIN industries i ON ei.industry_id = i.id
WHERE v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL
GROUP BY i.name
ORDER BY avg_salary DESC
LIMIT 20;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_3, connection)
df
```

```python
# текст запроса
query_7_4 = f'''
-- Средняя зарплата и количество вакансий по регионам
SELECT 
    a.name AS area,
    ROUND(AVG(
        (COALESCE(v.salary_from, 0) + COALESCE(v.salary_to, 0)) / 
        CASE 
            WHEN v.salary_from IS NOT NULL AND v.salary_to IS NOT NULL THEN 2 
            ELSE 1 
        END
    )) AS avg_salary,
    COUNT(v.id) AS vacancies_count
FROM vacancies v
JOIN areas a ON v.area_id = a.id
WHERE v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL
GROUP BY a.name
ORDER BY avg_salary DESC
LIMIT 20;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_7_4, connection)
df
```
